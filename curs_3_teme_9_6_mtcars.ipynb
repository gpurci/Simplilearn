```python
import pandas as pd
from sklearn import model_selection
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dataframe = pd.read_csv('/content/mtcars.csv')
dataframe
```

```python
dataframe.columns
```

```python
X = dataframe[['mpg', 'cyl', 'disp', 'drat', 'wt', 'qsec', 'hp', 'am', 'gear', 'carb']].values
Y = dataframe[['vs']].values
```

```python
seed = 7
num_trees = 30
```

```python
kfold = model_selection.KFold(n_splits=num_trees, random_state=seed, shuffle=True)
```

```python
for train_set, test_set in kfold.split(dataframe):
  print('train_set ', train_set)
  print('test_set ', test_set)
```

```python
from sklearn.ensemble import AdaBoostClassifier
model = AdaBoostClassifier(n_estimators=num_trees, random_state=seed)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
from sklearn import svm
from xgboost import XGBClassifier

model = XGBClassifier(n_estimators=num_trees, random_state=seed)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=num_trees)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=seed)
results = model_selection.cross_val_score(model, X, Y, cv=kfold)
print(results.mean())
```
